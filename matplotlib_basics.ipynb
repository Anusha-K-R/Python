### line plot

```python
year = [1980, 1990, 2000, 2010, 2020] # create a list of years
pop = [4.43, 5.32, 6.14, 6.92, 7.79]   # Population in billions
plt.plot(year, pop) # create a line plot
plt.xlabel('Year') # label x-axis
plt.ylabel('Population in Billions') # label y-axis
plt.title('World Population 1980 - 2020') # title of the plot
plt.yticks([0, 2, 4, 6, 8, 10],['0B', '2B', '4B', '6B', '8B', '10B'],) # set the y-axis scale
plt.show() # display the plot

# plt.clf() will clear the plot
```
